```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, recall_score, precision_score
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler
#reading the Obesity csv file
dataset = pd.read_csv('Obesity.csv')
#stores all the columns in x which are the features except the last column(label) 
X = dataset.drop(columns="NObeyesdad", axis=1)
#stores the last column (label)
y = dataset["NObeyesdad"]
#specify the columns with object data type (categorical columns only) to be selected 
categorical_cols = X.select_dtypes(include=['object']).columns.tolist()
# Define a column transformer to apply OneHotEncoding to categorical columns (drop first to avoid dummy variable trap)
column_transformer = ColumnTransformer(
    transformers=[
        ('ohe', OneHotEncoder(drop='first', sparse_output=False), categorical_cols)
    ],
    remainder='passthrough'
)
#apply the transformations in column_transformer to x (feature columns) 
# Apply the transformations in column_transformer to X (feature columns)
X_encoded = column_transformer.fit_transform(X)

# Standardize the numerical features using StandardScaler
scaler = StandardScaler()
X_encoded = scaler.fit_transform(X_encoded)

# LabelEncoder used to convert the target variable y (label) (NObeyesdad) to numerical value
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Split the dataset into training and testing sets (80% training, 20% testing)
xTrain, xTest, yTrain, yTest = train_test_split(X_encoded, y_encoded, test_size=0.2, random_state=42)
#intialize the knn model with 5 neighbors 
knn = KNeighborsClassifier(n_neighbors=5)
#trains the knn model using the training data (80% of the data) on the features and label
knn.fit(xTrain, yTrain)
#test the knn model using the testing data (20%) using features to see the label 
yPrediction = knn.predict(xTest)
#calculate the accuracy score 
accuracyScore = accuracy_score(yTest, yPrediction)
#calculate the confusion matrix 
confMatrix = confusion_matrix(yTest, yPrediction)
#calculate the recall  
recall = recall_score(yTest, yPrediction, average='macro')
#calculate the precision score 
precision = precision_score(yTest, yPrediction, average='macro')
#printing the results  
print("Result of Accuracy:", accuracyScore)
print("Result of Confusion Matrix:\n", confMatrix)
print("Result of Recall Score:", recall)
print("Result of Precision Score:", precision)
# print (dataset.columns)
# print (dataset.head())
```

Implementing the Decision Trees

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from sklearn.compose import ColumnTransformer

# Reading the Obesity CSV file
dataset = pd.read_csv('Obesity.csv')

# Stores all the columns in X which are the features (excluding the label column)
X = dataset.drop(columns="NObeyesdad", axis=1)

# Stores the label column (target variable)
y = dataset["NObeyesdad"]

# Select all the categorical columns from X
categorical_cols = X.select_dtypes(include=['object']).columns.tolist()

# Define a column transformer to apply OneHotEncoding to categorical columns (drop first to avoid dummy variable trap)
column_transformer = ColumnTransformer(
    transformers=[
        ('ohe', OneHotEncoder(drop='first', sparse_output=False), categorical_cols)
    ],
    remainder='passthrough'
)

# Apply the transformations to X to get the encoded features
X_encoded = column_transformer.fit_transform(X)

# Use LabelEncoder to convert the target variable (y) into numerical labels
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Split the dataset into training and testing sets (80% training, 20% testing)
xTrain, xTest, yTrain, yTest = train_test_split(X_encoded, y_encoded, test_size=0.2, random_state=42)

# Initialize the Decision Tree Classifier model
decision_tree = DecisionTreeClassifier(random_state=42)

# Train the decision tree model using the training data
decision_tree.fit(xTrain, yTrain)

# Predict the labels for the test data
yPrediction = decision_tree.predict(xTest)

#calculate the accuracy score 
accuracyScore = accuracy_score(yTest, yPrediction)
#calculate the confusion matrix 
confMatrix = confusion_matrix(yTest, yPrediction)
#calculate the recall  
recall = recall_score(yTest, yPrediction, average='macro')
#calculate the precision score 
precision = precision_score(yTest, yPrediction, average='macro')

#printing the results
print("Result of Accuracy:", accuracyScore)
print("Result of Confusion Matrix:\n", confMatrix)
print("Result of Recall Score:", recall)
print("Result of Precision Score:", precision)


# Extract feature names after encoding
encoded_feature_names = (
    column_transformer.named_transformers_['ohe'].get_feature_names_out(categorical_cols).tolist()
    + X.select_dtypes(exclude=['object']).columns.tolist()
)
#determine the size of the image as 20x10
plt.figure(figsize=(20, 10))
#pl
plot_tree(decision_tree, feature_names=encoded_feature_names, class_names=label_encoder.classes_, filled=True)
plt.show()
```
